# Zadanie 1
---
Użyj konstrukcji (__if...__), aby uzyskać poniższy efekt: 
- dla `a` równego __"dom"__  wyświetli się napis __"To jest słowo 'dom' "__ w przeciwnym wypadku __"nie znam tego wyrazu"__
- dla `b` równego __"ogród"__ wyświetli się napis __"To jest słowo 'ogród' "__  w przeciwnym wypadku __"to nie jest 'ogród' "__

````python
a = 'dom'
b = 'ogród'
````

````python
if a == 'dom':
    print("To jest slowo 'dom'")
else:
    print('nie znam tego wyrazy')
if b == 'ogród':
    print("To jest słowo 'ogród'")
else:
    print('to nie jest ogród')
````

# Zadanie 2
---
Przeiteruj listę `lista_dodatnich` (__for ... in ...__). Dla każedej szóstki wypisz: __"To jest przysta szóstka"__.  Dla pozostałych wartości podzielnych przez 2 wypisz __"Parzysta"__. Na koniec wypisz z użyciem polecenia __print__ największą liczbę z listy.

Oczekiwany wynik:
```
4 Parzysta
6 To jest parzysta szóstka
2 Parzysta
Wartość maksymalna listy to  9
```

````python
lista_dodatnich = [3, 4, 6, 2, 3, 9, 3]
````

````python
for element in lista_dodatnich:
    if element == 6:
        print('To jest parzysta szóstka')
    elif element % 2 == 0:
        print("Parzysta")

print(max(lista_dodatnich))
````

# Zadanie 4
---
Napisz funkcję __suma_ostatnich_dwoch(lista)__, która jako parametr przyjmuje listę `lista` i zwraca sumę ostatnich dwóch elementów.

Przykład użycia:
```
lista = [10,3,1,10,3]
suma_ostatnich_dwoch(lista)
13
```

````python
def suma_ostatnich_dwoch(lista: list()) -> int:
    return sum(lista[-2:])

lista_check = [10,3,1,10,7]
suma_ostatnich_dwoch(lista_check)
````

# Zadanie 5
---
1. Załaduj plik `'titanic_train.csv'` do zmiennej `titanic`. 
2. Uzupełnij brakujące wartości w kolumnie `Cabin` wartością `not_relevant`
3. Stwórz nową kolumnę `Expensive_ticket`, która zawiera wartość True dla `Fare` powyżej średniej `Fare` i False w przeciwnym wypadku.
4. Wybierz zbiór z kolumnami __PassengerId, Name, Fare, Expensive_ticket__, posortuj go według `Fare` (najdroższe bilety najwyżej) i ponownie zapisz go w tej samej zmiennej (`titanic`)
5. Wybierz rzeczywiste pierwsze __20__ wierszy i ponownie zapisz je w tej samej zmiennej (`titanic`)
6. Zapisz zbiór `titanic` jako plik CSV `titanic_zaliczenie.csv`

*ps. pliku `titanic_zaliczenie.csv` nie trzeba wysyłać mailem. Oceniony będzie tylko kod w tym Notebook'u.*

````python
import pandas as pd
import numpy as np
````

````python
titanic = pd.read_csv('./data/titanic_train.csv')
titanic
````

````python
titanic.loc[:,'Cabin'].fillna('not_relevant', inplace=True)
````

````python
titanic['Expensive_ticket'] = np.where(titanic.loc[:,'Fare'] > titanic.Fare.mean(), True, False)
titanic
````

````python
titanic = titanic.loc[:,['PassengerId', 'Name', 'Fare', 'Expensive_ticket']].sort_values(by='Fare', ascending=False)
````

````python
titanic = titanic.iloc[:20,:]
````

````python
titanic.to_csv('titanic_zaliczenie.csv', index=False)
````
