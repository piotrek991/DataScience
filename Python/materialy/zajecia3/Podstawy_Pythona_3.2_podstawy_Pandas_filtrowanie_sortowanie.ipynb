## Kilka filtrów jednocześnie

---
Każdy z filtrów (operacji logicznych) należy otoczyć nawiasami.

__Operacje logiczne (binarne)__  
  __&__ - AND  
  __|__ - OR

````python
# Osoby o nieznanym wieku z drugiej klasy
df[(df['Age'].isnull()) & (df.Pclass == 2)]
````

````python
# Starsi mężczyźni 
df[(df['Age']>=60) & (df['Sex']=='male')]
````

````python
# Kobety i dzieci
df[(df['Sex']=='female') | (df['Age']<18)]
````

````python
# Wiek powyżej mediany i pasażerowie drugiej klasy oraz wszyscy pasażerowie pierwszej klasy
df[((df['Age'] >= df['Age'].median()) & (df['Pclass'] == 2)) | (df['Pclass'] == 1)]
````

````python
# Osoby z tytułem 'Miss' z drugiej klasy, którzy weszli na podkład w 'C'
df[
    (df.Pclass == 2) & 
    (df.Embarked == 'C') &
    (df.Name.str.contains('Miss'))
]
````

## Negacja

````python
# Wszyscy pasażerowie, którzy nie mają tytułu 'Mrs'
df[~df['Name'].str.contains("Mrs")]
````

````python
# Wszyscy pasażerowie, którzy mają określoną nazwę kajuty 
df[~(df['Cabin'].isnull())]
````

# Ćwiczenia

````python
df = pd.read_csv('./data/titanic_train.csv')
````

### Wyszukaj pasażerów, którzy wsiedli w Queenstown (*Q*)

````python
df[df.Embarked == 'Q']
````

### Sprawdź czy są pasażerowie, którzy są w klasie `Pclass` 1 i zapłacili za bilet mniej, niż wynosi średnia cena biletu

````python
df[(df.Fare < df.Fare.mean()) & df.Pclass == 1]
````

### Sprawdź jaka jest średnia cena biletu w każdej z klas

````python
df.groupby('Pclass')[['Fare']].agg('mean')
````

### Sprawdź ile osób przeżyło w klasie pierwszej a ile w trzeciej

````python
df[df.Pclass.isin([1,3])].groupby('Pclass')['Survived'].sum()
````

### Sprawdź ilu pasażerów przeżyło, którzy wsiedli w porcie Southampton (*S*)

````python
df[df.Embarked == 'S'].loc[:,['Survived']].sum()
````

### Stwórz kolumnę `Expensive ticket`. Wpisz do niej wartości `1` jeżeli cena (`Fare`) biletu przekroczyła 31, w przeciwnym przypadku wpisz `0` 

````python
df['Expensive_ticket'] = np.where(df.Fare > 31,1,0)
df
````
