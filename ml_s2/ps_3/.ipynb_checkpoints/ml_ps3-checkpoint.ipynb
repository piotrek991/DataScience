```python
class AdalineSGD(object):

    def __init__(self, eta=0.01, epochs=50):
        self.eta = eta
        self.epochs = epochs

    def train(self, X, y, reinitialize_weights=True):

        if reinitialize_weights:
            self.w_ = np.zeros(1 + X.shape[1])
        self.cost_ = []

        for i in range(self.epochs):
            for xi, target in zip(X, y):
                output = self.net_input(xi)
                error = (target - output)
                self.w_[1:] += self.eta * xi.dot(error)
                self.w_[0] += self.eta * error

            cost = ((y - self.activation(X))**2).sum() / 2.0
            self.cost_.append(cost)
        return self

    def net_input(self, X):
        return np.dot(X, self.w_[1:]) + self.w_[0]

    def activation(self, X):
        return self.net_input(X)

    def predict(self, X):
        return np.where(self.activation(X) >= 0.0, 1, -1)
    
class Perceptron(object):

    def __init__(self, eta=0.01, epochs=50):
        self.eta = eta
        self.epochs = epochs

    def train(self, X, y):

        self.w_ = np.zeros(1 + X.shape[1])
        self.errors_ = []

        for _ in range(self.epochs):
            errors = 0
            for xi, target in zip(X, y):
                update = self.eta * (target - self.predict(xi))
                self.w_[1:] +=  update * xi
                self.w_[0] +=  update
                errors += int(update != 0.0)
            self.errors_.append(errors)
        return self

    def net_input(self, X):
        return np.dot(X, self.w_[1:]) + self.w_[0]

    def predict(self, X):
        return np.where(self.net_input(X) >= 0.0, 1, -1)
```

```python
dataset = np.random.rand(30,2)
dataset
```

```python
y_set = np.apply_along_axis(lambda x: 2*x[0]+x[1]+np.random.rand(),1,dataset)
```

```python
y_set
```

```python
%%time
ada = AdalineSGD(epochs=15, eta=0.01)

# shuffle data
np.random.seed(123)
idx = np.random.permutation(len(y_set))
X_shuffled, y_shuffled =  dataset[idx], y_set[idx]
# train and adaline and plot decision regions
ada.train(X_shuffled, y_shuffled)
#plot_decison_regions(X_shuffled, y_shuffled, clf=ada)
plt.plot([i for i in range(30)]
         , ada.net_input(X_shuffled)
        , label = 'predicted')
plt.plot([i for i in range(30)]
            , y_shuffled
            , c='red'
            , label='actual')
plt.legend(loc="upper left")
plt.title('Adaline')
plt.xlabel('number of observation')
plt.ylabel('func(x1,x2)')
plt.show()

plt.plot(range(1, len(ada.cost_)+1), ada.cost_, marker='o')
plt.xlabel('Iterations')
plt.ylabel('Sum-squared-error')
plt.show()
print(f'calculated wages : {ada.w_}')
```

```python
dane_ex3 = pd.read_csv('ps3_dane1.txt',sep='\t')
dane_ex3
```

```python
x_data = dane_ex3.loc[:, ['x1', 'x2']].values
x_data
```

```python
X_std = np.copy(x_data)
X_std[:,0] = (x_data[:,0] - x_data[:,0].mean()) / x_data[:,0].std()
X_std[:,1] = (x_data[:,1] - x_data[:,1].mean()) / x_data[:,1].std()
X_std
```

```python
y_data = np.where(dane_ex3.loc[:, 'klasa'].values == 1, 1, -1)
y_data
```

```python
%%time
ada = AdalineSGD(epochs=15, eta=0.01)
ppn = Perceptron(epochs=25, eta=0.01)

# shuffle data
np.random.seed(123)
idx = np.random.permutation(len(y_data))
X_shuffled, y_shuffled =  X_std[idx], y_data[idx]
```

```python
ada.train(X_shuffled, y_shuffled)
ppn.train(X_shuffled, y_shuffled)

plot_decision_regions(X_shuffled, y_shuffled, clf=ada)
plt.title('Adaline - Gradient Descent')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()

plt.plot(range(1, len(ada.cost_)+1), ada.cost_, marker='o')
plt.xlabel('Iterations')
plt.ylabel('Sum-squared-error Adaline')
plt.show()
```
