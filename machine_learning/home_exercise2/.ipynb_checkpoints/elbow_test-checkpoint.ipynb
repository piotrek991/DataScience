```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn import metrics
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_selection import RFE
from mrmr import mrmr_classif
import sklearn_relief as sr
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
import time
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_dlbcl_train = pd.read_csv('dlbcl_training.csv')
data_dlbcl_test = pd.read_csv('dlbcl_testing.csv', delimiter=';')

data_prostate_train = pd.read_csv('prostate_training.csv')
data_prostate_test = pd.read_csv('prostate_testing.csv')

#x_train, y_train = data_dlbcl_train.iloc[:,:-1], data_dlbcl_train.iloc[:,-1]
#x_test, y_test = data_dlbcl_test.iloc[:,:-1], data_dlbcl_test.iloc[:,-1]

knn = KNeighborsClassifier(n_neighbors=5, metric='euclidean')

clf = DecisionTreeClassifier()

svc = SVC(gamma='auto', probability=True, kernel='linear')
```

```python
x_train, y_train = data_prostate_train.iloc[:,:-1], data_prostate_train.iloc[:,-1]
x_test, y_test = data_prostate_test.iloc[:,:-1], data_prostate_test.iloc[:,-1]
```

```python
##tree classificaton
#dlbcl
x_train, y_train = data_dlbcl_train.iloc[:,:-1], data_dlbcl_train.iloc[:,-1]
x_test, y_test = data_dlbcl_test.iloc[:,:-1], data_dlbcl_test.iloc[:,-1]
#prostate testing
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x_train, y_train = data_prostate_train.iloc[:,:-1], data_prostate_train.iloc[:,-1]
x_test, y_test = data_prostate_test.iloc[:,:-1], data_prostate_test.iloc[:,-1]

le.fit(y_train)
y_train = le.transform(y_train)
y_test = le.transform(y_test)

clf.fit(x_train, y_train)

f1_scores = list()
acc_scores = list()
features = list()

num_features_start = x_train.shape[1]

while x_train.shape[1] > 0:
    print(f'Training model on the {x_train.shape[1]} most important features')
    feature_importance = clf.feature_importances_
    y_pred = clf.predict(x_test)
    f1 = metrics.f1_score(y_test, y_pred)
    acc = metrics.accuracy_score(y_test, y_pred)
    acc_scores.append(acc)
    f1_scores.append(f1)
    features.append(x_train.columns)
    
    if x_train.shape[1] == 1:
        break
    
    least_important_idx = feature_importance.argmin()
    x_train = x_train.drop(x_train.columns[least_important_idx], axis=1)
    x_test = x_test.drop(x_test.columns[least_important_idx], axis=1)

    clf.fit(x_train,y_train)
```

```python
#mrmr method
# x_train, y_train = data_dlbcl_train.iloc[:,:-1], data_dlbcl_train.iloc[:,-1]
# x_test, y_test = data_dlbcl_test.iloc[:,:-1], data_dlbcl_test.iloc[:,-1]
###########
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x_train, y_train = data_prostate_train.iloc[:,:-1], data_prostate_train.iloc[:,-1]
x_test, y_test = data_prostate_test.iloc[:,:-1], data_prostate_test.iloc[:,-1]

le.fit(y_train)
y_train = le.transform(y_train)
y_test = le.transform(y_test)

final_data = list()
feature_list = list()

for i in range(1,51):
    print(f'here calculating model with only {i} features selected')

    selected_features = mrmr_classif(X = x_train, y = y_train, K=i)
    x_train_after_sel = x_train.loc[:, selected_features].to_numpy()
    x_test_after_sel = x_test.loc[:, selected_features].to_numpy()

    clf.fit(x_train_after_sel,y_train)
    knn.fit(x_train_after_sel,y_train)
    svc.fit(x_train_after_sel,y_train)
    y_pred_clf = clf.predict(x_test_after_sel)
    y_pred_knn = knn.predict(x_test_after_sel)
    y_pred_svc = svc.predict(x_test_after_sel)
    
    acc_score_clf = metrics.accuracy_score(y_test, y_pred_clf)
    acc_score_knn = metrics.accuracy_score(y_test, y_pred_knn)
    acc_score_svc = metrics.accuracy_score(y_test, y_pred_svc)

    final_data.append(['CLF','MRMR', i, acc_score_clf])
    final_data.append(['knn','MRMR', i, acc_score_knn])
    final_data.append(['svc','MRMR', i, acc_score_svc])
    
    feature_list.append(selected_features)

final_df = pd.DataFrame(columns = ['Model', 'Selection Method', 'Number of features', 'Accuracy'], data = final_data)
```

```python
#RELIEFF method
x_train, y_train = data_dlbcl_train.iloc[:,:-1], data_dlbcl_train.iloc[:,-1]
x_test, y_test = data_dlbcl_test.iloc[:,:-1], data_dlbcl_test.iloc[:,-1]
###########
# from sklearn.preprocessing import LabelEncoder
# le = LabelEncoder()
# x_train, y_train = data_prostate_train.iloc[:,:-1], data_prostate_train.iloc[:,-1]
# x_test, y_test = data_prostate_test.iloc[:,:-1], data_prostate_test.iloc[:,-1]

# le.fit(y_train)
# y_train = le.transform(y_train)
# y_test = le.transform(y_test)

final_data = list()
feature_list = list()

for i in range(1,51):
    print(f'here calculating model with only {i} features selected')
    
    scaler = sr.ReliefF(n_features = i)
    scaler.fit(x_train.to_numpy(), y_train.to_numpy())
    
    x_train_after_sel = x_train.loc[:, selected_features].to_numpy()
    x_test_after_sel = x_test.loc[:, selected_features].to_numpy()

    clf.fit(x_train_after_sel,y_train)
    knn.fit(x_train_after_sel,y_train)
    svc.fit(x_train_after_sel,y_train)
    y_pred_clf = clf.predict(x_test_after_sel)
    y_pred_knn = knn.predict(x_test_after_sel)
    y_pred_svc = svc.predict(x_test_after_sel)
    
    acc_score_clf = metrics.accuracy_score(y_test, y_pred_clf)
    acc_score_knn = metrics.accuracy_score(y_test, y_pred_knn)
    acc_score_svc = metrics.accuracy_score(y_test, y_pred_svc)

    final_data.append(['CLF','MRMR', i, acc_score_clf])
    final_data.append(['knn','MRMR', i, acc_score_knn])
    final_data.append(['svc','MRMR', i, acc_score_svc])
    
    feature_list.append(selected_features)

final_df = pd.DataFrame(columns = ['Model', 'Selection Method', 'Number of features', 'Accuracy'], data = final_data)
```

```python
#RFEfmethod method
# x_train, y_train = data_dlbcl_train.iloc[:,:-1], data_dlbcl_train.iloc[:,-1]
# x_test, y_test = data_dlbcl_test.iloc[:,:-1], data_dlbcl_test.iloc[:,-1]
####
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x_train, y_train = data_prostate_train.iloc[:,:-1], data_prostate_train.iloc[:,-1]
x_test, y_test = data_prostate_test.iloc[:,:-1], data_prostate_test.iloc[:,-1]

le.fit(y_train)
y_train = le.transform(y_train)
y_test = le.transform(y_test)

final_data = list()
feature_list = list()

for i in range(1,51):
    print(f'here calculating model with only {i} features selected')
    
    selector = RFE(clf, n_features_to_select=i, step=1)
    selector.fit(x_train, y_train)
    x_train_after_sel = selector.transform(x_train)
    x_test_after_sel = selector.transform(x_test)
    
    knn.fit(x_train_after_sel,y_train)
    clf.fit(x_train_after_sel,y_train)
    y_pred = knn.predict(x_test_after_sel)
    acc_score = metrics.accuracy_score(y_test, y_pred)

    final_data.append(['knn','ReliefF', i, acc_score])
    feature_list.append([col for col,rank in zip(x_train.columns,selector.ranking_) if rank == 1])

final_df = pd.DataFrame(columns = ['Model', 'Selection Method', 'Number of features', 'Accuracy'], data = final_data)
```
